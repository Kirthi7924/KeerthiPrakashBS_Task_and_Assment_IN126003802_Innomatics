```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

c = {}

for i in emails:
    d = i.split("@")[1]
    if d not in c:
       c[d] = 1
    else:
       c[d] += 1
t = len(emails)

for d, c in c.items():
    p = (c / t) * 100
    print(f"{d}: {int(p)}%")
```
