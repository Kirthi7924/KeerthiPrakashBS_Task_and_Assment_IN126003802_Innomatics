```python
def rainfall_check(rainfall):
    total = 0
    for i in rainfall:
        total+=i
    average = total/len(rainfall)
    if average>=50:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    return average,status
rainfall = [int(i) for i in input("Enter rainfall:").split()]
avg,sts = rainfall_check(rainfall)
print("Average rainfall:",avg)
print("Status:",sts)
```
