```python
def post_analyser(likes):
    total = 0
    for i in likes:
        total += i
    if total>=1000:
        status="Viral Post"
    else:
        status="Normal Engagement"
    print("Total likes:",total)
    print(status)
likes = [int(i) for i in input("Enter likes count:").split()]
post_analyser(likes)
```
