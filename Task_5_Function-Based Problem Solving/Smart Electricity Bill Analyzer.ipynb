```python
def electricity_bill(unit):
    total = 0
    if unit<=100:
        total = unit*3
    elif unit<=200:
        total = 100*3
        total+=(unit-100)*5
    else:
        total = 100*3
        total+=100*5
        total+=(unit-200)*7

    if total<=500:
        status = "Low Usage"
    elif total<=1500:
        status = "Moderate usage"
    else:
        status = "High usage"
    return status
unit = int(input("Enter units used:"))
print(electricity_bill(unit))
```
