```python
def duplicate_account(account):
    l = len(account)
    n = len(set(account))
    if l==n:
        return True
    else:
        return False
account = [i for i in input("Enter usernames:").split()]
if duplicate_account(account):
    print("Duplicate Accounts Found: No")
else:
    print("Duplicate Accounts Found: Yes")
```
