```python
def student_eligibility(attendance):
    total = 0
    for i in attendance:
        if i == "p":
            total+=1
    percentage = total/len(attendance)*100
    if percentage>=75:
        status="Eligible"
    else:
        status="Not eligible"
    return percentage,status
attendance = [i.lower() for i in input("Enter attendance(A/P):").split()]
per,sts = student_eligibility(attendance)
print("Attendance percentage:",per)
print("Exam Eligibility:",sts) 
```
