```python
def price_filter(price):
    prm_crp=[]
    reg_crp=[]
    for i in price:
        if i<2000:
            reg_crp.append(i)
        else:
            prm_crp.append(i)
    return prm_crp
price = [int(i) for i in input("Enter crop price:").split()]
print("Premium crops:",price_filter(price))
```
