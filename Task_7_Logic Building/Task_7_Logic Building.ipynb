Smart Light Controller

```python
class light_control:
    def __init__(self,name):
        self.name = name
        self.status = "OFF"
    def turn_on(self):
        self.status = "ON"
        print(self.name,"is turned on")
    def turn_off(self):
        self.status = "OFF"
        print(self.name,"is turned off")
    def show_status(self):
        print(f"{self.name} is {self.status}")
bedroom_light = light_control("Beedroom Light")
bedroom_light.turn_on()
bedroom_light.show_status()
bedroom_light.turn_off()
bedroom_light.show_status()
```

Employee ID Card System

```python
class ID_card:
    def __init__(self,name,ID,dep):
        self.name = name
        self.ID = ID
        self.dep = dep
    def print_id(self):
        print("Employee ID Card")
        print("Name:",self.name)
        print("ID:",self.ID)
        print("Department:",self.dep)
employee = ID_card("Keerthi","IN126003802","Intern")
employee.print_id()
```

Mobile Contact Record

```python
class Contact_record:
    def __init__(self,name,number):
        self.name = name
        self.number = number
    def save_contact(self):
        print("Contact saved")
        print("Name:",self.name)
        print("Phone:",self.number)
con = Contact_record("Keerthi",8088205362)
con.save_contact()
```
