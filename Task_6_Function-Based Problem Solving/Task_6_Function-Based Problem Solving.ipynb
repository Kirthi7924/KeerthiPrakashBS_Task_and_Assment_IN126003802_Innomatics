Problem Statement 1:Smart Parking Lot Management System

```python
def parking(capacity,log):
    count = 0
    for i in log:
        if i == "in" and count >= capacity:
            status = "Parking out of range"
            return count, status
        elif i == "in" and count < capacity:
            count += 1
        else:
            count -= 1
    if count < capacity:
        status = "Available"
    else:
        status = "Not Available"
    return count, status
capacity = int(input("Enter parking capacity:"))
log = [ i.lower() for i in input("Enter vehical logs:").split()]
count, sts = parking(capacity,log)

print("Vehicals parked:",count)
print("Parking Status:",sts)
```

Problem Statement 2:Online Food Delivery Time Estimator

```python
def time_calculate(distance, traffic, weather):
    time = distance/30*60
    if traffic == "high":
        time += 20
    elif traffic == "medium":
        time += 15
    elif traffic == "low":
        time += 10
    if weather == "rainy":
        time += 15
    elif weather == "sunny":
        time
    return time
distance = int(input("Enter distance:"))
traffic = input("Enter Traffic condition(high/low/medium/no):").lower()
weather = input("Enter weather condition(rainy/suuny/normal):").lower()
print(f"Estimated time delivery: {time_calculate(distance,traffic,weather)}minutes")
```

Problem Statement 3:Movie Theatre Seat Occupancy 

```python
def seat_occupancy(seats,booked):
    occupancy = len(booked)/seats*100
    if occupancy>=70:
        status = "Almost full"
    elif occupancy >= 50:
        status = "Fast filling"
    else:
        stauts = "Seats available"
    return occupancy,status
seats = int(input("Enter the total seats:"))
booked = [int(i) for i in input("Enter booked seats:").split()]
occ, sts = seat_occupancy(seats,booked)
print(f"Occupancy:{occ}%")
print(f"Show status:{sts}")
```

Problem Statement 4:Cloud Server Load Classification System

```python
def server_load(readings):
    total = 0
    for i in readings:
        total += i
    average = total/len(readings)
    if average <50:
        status = "Normal"
    elif average < 80:
        status = "Warning"
    else:
        status = "Critical"
    return average,status
readings = [int(i) for i in input("Enter CPU Readings:").split()]
avg,sts = server_load(readings)
print("Average CPU Load:",avg)
print("Server status:",sts)
```

Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def resource_usage(usage):
    over = 0
    used = ""
    total = 0
    for i, j in usage.items():
        if j > over:
            used = i
            over = j
        total += j
    if total>15:
        status = "Yes"
    else:
        status = "No"
    return used,status

usage = {}
n = int(input("Enter number of resources:"))
for i in range(n):
    res = input("Enter resource name: ")
    time = int(input("Enter usage time: "))
    usage[res] = time
res,sts = resource_usage(usage)
print("Overused Resources:",res)
print("Energy Alert:",sts)
```

Problem Statement 6: Online Event Registration Capacity Controller


```python
def registraction(cap,reg):
    if reg > cap:
        confirmed = cap
        waitlisted = reg-cap
        status = "Closed"
    else:
        confirmed = reg
        waitlisted = 0
        status = "Seats available"
    return confirmed,waitlisted,status
capacity = int(input("Enter event capacity:"))
registered = int(input("Enter the no of registered:"))
con,wai,sts = registraction(capacity,registered)
print("Confirmed Registrations:",con)
print("Waitlisted Users:",wai)
print("Registration Status:", sts)  
```
