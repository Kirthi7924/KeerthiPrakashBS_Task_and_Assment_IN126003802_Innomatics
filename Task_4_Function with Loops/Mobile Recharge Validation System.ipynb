```python
def recharge(a):
    p = [199, 299, 399, 599]
    if a < 50:
        print("Recharge amount should be more than ₹50.")
        return False
    if a not in p:
        print("Invalid plan")
        return False
    return True
while True:
    a = int(input("Enter recharge amount: ₹"))
    if recharge(a):
        print("Recharge successful!")
        break
    else:
        print("Please try again.\n")
```
